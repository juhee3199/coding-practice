##### 구현 유형의 문제
: 풀이를 떠올리는 것은 쉽지만 소스코드로 옮기기 어려운 문제
- 시뮬레이션과 완전탐색으로 해결

```python
# 동, 북, 서, 남의 방향벡터
dx = [0, -1, 0, 1] # 행
dy = [1, 0, -1, 0] # 열
# ex) '동'벡터는 행은 그대로 열은 오른쪽으로.

# 현재 위치
x, y = 2,2

# 현재 위치를 기준으로 동, 북, 서, 남의 위치가 출력됌
for i in range(4):
    # 다음 위치
    nx = x+dx[i]
    ny = y+dy[i]
    print(nx, ny)
```

- <문제> 시뮬레이션: 상하좌우

```python
n = int(input())
plans = input().split()

x = 1 # 행
y = 1 # 열

# L, R, U, D 에 따른 이동 방향
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L','R','U','D']

for plan in plans:
    # 이동 후 좌표 구하기
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    
    # 공간을 벗어나는 경우 무시
    if nx<1 or ny<1 or nx>n or ny>n:
        continue
    
    # 이동 수행
    x, y = nx, ny

print(x, y)
```

- <문제> 완전 탐색 : 시각 문제

```python
h = int(input())

count=0
for i in range(h+1):
    for j in range(60):
        for k in range(60):
            # 매 시각 안에 '3'이 포함되어 있다면 카운트 증가
            if '3' in str(i) + str(j) +  str(k):
                count+=1
                
print(count)
```

- < 문제 > 왕실의 나이트

```python
input_data = input()
row = int(input_data[1])
col = int(ord(input_data[0])) - int(ord('a')) + 1  #ord(): 유니코드 값을 반환

# 나이트가 이동할 수 있는 8가지 방향 정의
steps = [(-1,-2), (1, -2), (-1,2), (1,2), (-2, -1), (-2,1), (2, -1), (2,1)]
   
result=0
for step in steps:
    next_row = row+step[0]
    next_column = col + step[1]
    
    # 이동 가능하다면 카운트 증가
    if next_row>=1 and next_row<=8 and next_column>=1 and next_column<=8:
        result+=1

print(result)
```

- <문제> 문자열 재정렬

```python
data = input()

result=[]
value=0

for x in data:
    if x.isalpha():
        result.append(x)
        
    else:
        value += int(x)

result.sort()

# 숫자가 존재할 경우 맨 뒤에 삽입
if value!=0: 
    result.append(str(value))

print(''.join(result))
```
